# model details

```python
N = 1
y = 'is_turnpt_downward'
```

```python
from prepare_X import get_featrues, specs

df_features = get_featrues(specs)
df_features.index = df_features.index.date

df_features.head()

df_features.shape
```

```python
from turnpt_analysis import get_df_turnpt_measures


df_y = get_df_turnpt_measures(N).fillna(0)
df_y.index = df_y.tx_datetime

df_chosen_y = df_y[[y]]

df_chosen_y.head()
```

```python
df = df_chosen_y.join(df_features).dropna()
```

```python
import numpy as np
from catboost import CatBoostClassifier, Pool
from catboost import CatBoostRegressor
```

```python
num_train = int(df.shape[0] * 0.75)
y = df[y] 
y_train = y[: num_train]
y_test = y[num_train:]

X = df.iloc[:, 1:]
X_train = X.iloc[:num_train, :]
X_test = X.iloc[num_train:, :]
```

```python
train_pool = Pool(X_train, y_train, weight=y_train.abs()+1)
test_pool = Pool(X_test, y_test, weight=y_test.abs()+1)
```

```python
import pickle, io
```

```python
with io.open('models/model_1_is_turnpt_upward.pickle', 'rb') as f:
    model = pickle.load(f)
```

```python
# best_iter = np.argmax(m['Precision'])
# print(best_iter)
```

```python
# model.plot_tree(best_iter)
```

```python
m = model.eval_metrics(test_pool, ['Precision', 'Recall'])
best_iter = np.argmax(m['Precision'])
m['Precision'][best_iter], m['Recall'][best_iter]
```
